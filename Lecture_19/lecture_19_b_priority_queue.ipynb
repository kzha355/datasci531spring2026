### **Priority Queues with Binary Heaps**

<font size = "4">

- A **priority queue** is a variant of a queue where the order of stored items is determined by their *priority*

- In effect, when we add an item, it can be moved right to the front if it has the highest priority.

- We can implement a priority queue using lists and sorting algorithms:

    - Inserting into a list is $\mathcal{O}(n)$
    - Sorting a list (initializing a priority queue) is $\mathcal{O}(n\log n)$

- Priority queues can be more efficiently implemented using a **binary heap** data structure.

    - Enqueuing and dequeuing is $\mathcal{O}(\log n)$

    - "Heapifying" a list (initializing a binary heap) is $\mathcal{O}(n)$

- A **min heap** has the smallest key at the front of the queue. A **max heap** has the largest key at the front of the queue.

- We will focus on a **min heap**.


#### **Binary Heap Implementation**

<font size = "4">

- **Key property:** The logical structure of a binary heap is a binary tree, but is implemented as an **array** (i.e., a Python list).

- Specifically, the logical structure is a **complete binary tree**. Each level has all of its nodes, with the possible exception of the last level.

<font size = "4">

- The levels are filled in "left-to-right". We can actually implement this as a single list/array.

    - Root has index 0.
    - Root's left child: index 1. Root's right child: index 2.
    - Continue on, going down levels, left-to-right

- The nodes in level $\ell \geq 0$ have indices $2^\ell - 1 \leq i < 2^{\ell+1} - 1$

- If a node has index $p$, then its children have indices $2p + 1$ (left child) and $2p + 2$ (right child).

- So if an item is placed in index $n$ in the array, its parent has the index $\lfloor \frac{n-1}{2}\rfloor$ (floor).

- This can be computed using integer division as `(n-1) // 2`

#### **Heap Order Property**

<font size = "4">

- For every node $x$ with parent $p$, the key in $p$ is smaller than or equal to the key in $x$.

- This is really the **min heap property**

#### **Binary Heap Implementation**

````python
class BinaryHeap:
    def __init__(self):
        self._heap = []

    def _perc_up(self, i):
        while (i - 1) // 2 >= 0:
            p = (i - 1) // 2
            if self._heap[i] < self._heap[p]:
                self._heap[i],self._heap[p] = (
                    self._heap[p], self._heap[i]
                )
            else:
                break
            i = p

    def insert(self, item):
        self._heap.append(item)
        self._perc_up(len(self._heap) - 1)

    def _get_min_child(self, i):
        if 2 * i + 2 > len(self._heap) - 1: # no right child
            return 2 * i + 1
        if self._heap[2 * i + 1] < self._heap[2 * i + 2]:
            # tie -> take right_child
            return 2 * i + 1
        return 2 * i + 2
    
    def _perc_down(self, i):
        while 2 * i + 1 < len(self._heap):
            sm_child = self._get_min_child(i)
            if self._heap[i] > self._heap[sm_child]:
                self._heap[i], self._heap[sm_child] = (
                    self._heap[sm_child], self._heap[i]
                )
            else:
                break
            i = sm_child

    def delete(self):
        # swap, pop, percolate down
        self._heap[0], self._heap[-1] = self._heap[-1], self._heap[0]
        result = self._heap.pop()
        self._perc_down(0)
        return result

    def heapify(self, not_a_heap):
        # make a copy of the input list
        self._heap = not_a_heap[:]
        # left-most node in second-lowest level
        i = len(self._heap) // 2 - 1
        while i >= 0:
            self._perc_down(i)
            i = i - 1
    
    def is_empty(self):
        return not bool(self._heap)

    def __len__(self):
        return len(self._heap)

    def __str__(self):
        return str(self._heap)
````

````python
my_heap = BinaryHeap()
my_heap.insert(5)
my_heap.insert(7)
my_heap.insert(3)
my_heap.insert(11)

# Smallest item has "highest priority"
# is the first to leave the queue

print(my_heap.delete())
print(my_heap.delete())
print(my_heap.delete())
print(my_heap.delete())
````

#### **Binary Heap Operations**

<font size = "4">

*Insert method*

1. Append new item to the end of the list. $\mathcal{O}(1)$, but destroys heap property.

2. "Percolate" the item up the tree, swapping with parent node when the parent has a larger value.

*Delete method*

1. Want to remove the root (minimum item). This is at beginning of list, so swap with end and pop.

2. To restore min heap property, "percolate" the node down the tree, choosing the minimum of the two children.

```python
        sm_child = self._get_min_child(i)

            )
        else:
            break

    self._perc_down(0)
    return result
```

<font size = "4">

Both inserting and deleting (queuing and enqueuing) are $\mathcal{O}(\log n)$ operations.

<font size = "4">

*Heapify method*

- This method takes a list of unordered keys and rearranges them into a heap structure.

- Could try inserting one at a time. This would be $n$ insertions at $\mathcal{O}(\log n)$ time each. So $\mathcal{O}(n \log n)$ to build a heap this way. (This is called Williams' method)

- Can use another method to get $\mathcal{O}(n)$ cost.

    1. Initialize the binary tree in the order of the list, ignoring the heap property.

    2. "Heapify" the subtrees with roots on the 2nd-to-last level by percolating down. (Going right-to-left).

    3. Then "heapify" the subtrees with roots on the 3rd-to-last level.

    4. Continue until we reach the root, then "heapify" the entire tree by percolating down.

- This algorithm takes $\mathcal{O}(n)$ time. 

````python
a_heap = BinaryHeap()
a_heap.heapify([9, 6, 5, 2, 3])

print(a_heap._heap)
print()
while not a_heap.is_empty():
    print(a_heap.delete())
````
