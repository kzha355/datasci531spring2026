- Instead of storing the data as rows in a Pandas DataFrame, you will implement your own data structure, a *VoterRegistry* class. 

- Instead of storing data in rows, a VoterRegistry stores multiple Voter objects, which is implemented in the next cell.

```python
class Voter:
    def __init__(self, voter_id, age, county, party, income_bracket, turnout_history):
        self.voter_id = voter_id
        self.age = age
        self.county = county
        self.party = party
        self.income_bracket = income_bracket
        self.turnout_history = turnout_history  # list of 0/1

    def turnout_rate(self):
        if not self.turnout_history:
            return 0.0
        return sum(self.turnout_history) / len(self.turnout_history)

    def __repr__(self):
        return (
            f"Voter(id={self.voter_id}, age={self.age}, "
            f"county='{self.county}', party='{self.party}', "
            f"income='{self.income_bracket}')"
        )

# Example use for one data point from a generated dataset
data = generate_dataset(n=100)
single_data_pt = data[0]
turn_out = [single_data_pt["t_2016"], single_data_pt["t_2018"],
    single_data_pt["t_2020"], single_data_pt["t_2022"]]

voter = Voter(single_data_pt["voter_id"],
            single_data_pt["age"],
            single_data_pt["county"],
            single_data_pt["party"],
            single_data_pt["income_bracket"],
            turn_out)

print(voter)
print(voter.turnout_rate())
```
