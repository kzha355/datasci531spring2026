- Instead of storing the data as rows in a Pandas DataFrame, you will implement your own data structure, a *VoterRegistry* class. 

- Instead of storing data in rows, a VoterRegistry stores multiple Voter objects, which is implemented in the next cell.

```python
class Voter:
    def __init__(self, voter_id, age, county, party, income_bracket, turnout_history):
        self.voter_id = voter_id
        self.age = age
        self.county = county
        self.party = party
        self.income_bracket = income_bracket
        self.turnout_history = turnout_history  # list of 0/1

    def turnout_rate(self):
        if not self.turnout_history:
            return 0.0
        return sum(self.turnout_history) / len(self.turnout_history)

    def __repr__(self):
        return (
            f"Voter(id={self.voter_id}, age={self.age}, "
            f"county='{self.county}', party='{self.party}', "
            f"income='{self.income_bracket}')"
        )

# Example use for one data point from a generated dataset
data = generate_dataset(n=100)
single_data_pt = data[0]
turn_out = [single_data_pt["t_2016"], single_data_pt["t_2018"],
    single_data_pt["t_2020"], single_data_pt["t_2022"]]

voter = Voter(single_data_pt["voter_id"],
            single_data_pt["age"],
            single_data_pt["county"],
            single_data_pt["party"],
            single_data_pt["income_bracket"],
            turn_out)

print(voter)
print(voter.turnout_rate())
```

## Implementation: VoterRegistry

<font size = "4">

### Data Structure Choice
I implemented `VoterRegistry` using a linked list, where each `Node` stores one `Voter` object and a pointer to the next node. This design was chosen for its simplicity: O(1) insertion at the head and no hash collision handling required.

<font size = "4">

### `Node`
A class that stores `Voter` object and a `next` pointer to the following node. 

```python
class Node:
    def __init__(self, voter):
        self.voter = voter
        self.next = None
```

<font size = "4">

### `VoterRegistry`
It stores all `Voter` objects in a linked list, where each `Voter` is wrapped in a `Node` and connected to the next one via a pointer. Below I will introduce the function within the class and how they enable sorting and searching.

- `dict_to_voter()`: Converts a dictionary from `generate_dataset()` into a `Voter` object, and package the four turnout fields into a list.

- `add()`: Inserts a new `Voter` at the head of the list by creating a new `Node`, pointing its next to the current head, then updating the head to the new node.

- `conv_to_list()`: Traverses the linked list, collecting each `Voter` object into a Python list. And this function is called within the class to enable sorting operations.

- `conv_from_list()`: Rebuilds the linked list from a sorted list.

- `sort()`: Sorts the registry by any voter attribute using merge sort (O(n log n)). Converts to list, sorts, rebuilds linked list.

- `search()`: Traverses the list and returns `(index, Voter)` pairs for all voters based on the given condition.

- `__repr__()`: Defines how the registry is displayed when printed. Each voter is shown with their attributes and `turnout_rate`, since `turnout_rate` is not stored as a field but we need it to visually verify that sorting by `turnout_rate` is correct.

#### `income_bracket` Ordering
`income_bracket` is a categorical string. To sort it by actual income order rather than alphabetical order, it is mapped to integers: 
- "<50K"→0
-"50-100K"→1
-">100K"→2


#### Sorting: Merge Sort
Merge sort was chosen because it works on lists and has O(n log n) performance.


```python
class VoterRegistry:
    def __init__(self, data):
        self.head = None
        self.size = 0
        if data:
            for d in data:
                self.add(self.dict_to_voter(d))
    def dict_to_voter(self, d):
        turnout_history = [d["t_2016"], d["t_2018"], d["t_2020"], d["t_2022"]]
        return Voter(d["voter_id"], d["age"], d["county"], d["party"], d["income_bracket"], turnout_history)
    
    def add(self, voter):
        temp = Node(voter)
        temp.next = self.head 
        self.head = temp
        self.size += 1
    
    def conv_to_list(self):
        result = []
        current = self.head
        while current:
            result.append(current.voter)
            current = current.next
        return result
    
    def conv_from_list(self, voters):
        self.head = None
        self.size = 0
        for v in voters:
            self.add(v)
    
    def sort(self, indice):
         voters = self.conv_to_list()
         sorted_voters = self.merge_sort(voters, indice)
         self.conv_from_list(sorted_voters)

    var_dict = {
        "voter_id": lambda v: v.voter_id,
        "age": lambda v: v.age,
        "county": lambda v: v.county,
        "party": lambda v: v.party,
        "income_bracket": lambda v: {"<50K": 0, "50-100K": 1, ">100K": 2}[v.income_bracket],
        "t_2016": lambda v: v.turnout_history[0],
        "t_2018": lambda v: v.turnout_history[1],
        "t_2020": lambda v: v.turnout_history[2],
        "t_2022": lambda v: v.turnout_history[3],
        "turnout_rate": lambda v: v.turnout_rate(),
    }

    def merge_sort(self, lst, indice):
        if len(lst) <= 1:
            return lst

        mid = len(lst) // 2
        left_half = lst[:mid]
        right_half = lst[mid:]

        # recursive call to each half
        left_half = self.merge_sort(left_half, indice)
        right_half = self.merge_sort(right_half, indice)
        
        ind = self.var_dict[indice]
        i, j, k = 0, 0, 0
        while i < len(left_half) and j < len(right_half):
            if ind(left_half[i]) <= ind(right_half[j]):
                lst[k] = left_half[i]
                i = i + 1
            else:
                lst[k] = right_half[j]
                j = j + 1
            k = k + 1

        while i < len(left_half):
            lst[k] = left_half[i]
            i = i + 1
            k = k + 1

        while j < len(right_half):
            lst[k] = right_half[j]
            j = j + 1
            k = k + 1
        return lst
    
    def search(self, condition):
        results = []
        current = self.head
        index = 0
        while current is not None:
            if condition(current.voter):
                results.append((index, current.voter))
            current = current.next
            index += 1
        return results

    def __repr__(self):
        lines = [f"VoterRegistry ({self.size} voters)"]
        for i, v in enumerate(self.conv_to_list()):
            lines.append(f" [{i}] {v} | turnout_rate={v.turnout_rate()}")
        return "\n".join(lines)
```

### Benchmarking

I tested sorting and search performance across data sizes of n = 10, 100, 1000, 10000.

```python
import time

# varying n
for n in [10, 100, 1000, 10000]:
    data = generate_dataset(n=n)
    registry = VoterRegistry(data)

    t0 = time.time()
    registry.sort("voter_id")
    sort_time = time.time() - t0

    t0 = time.time()
    results = registry.search(lambda v: v.age > 50)
    search_time = time.time() - t0

    print(f"n={n} | sort: {sort_time:.6f}s | search: {search_time:.6f}s")
```

<font size = "4">

- Sort time grows roughly as O(n log n): from n=1000 to n=10000 (10x data), sort time increases ~13x.

- Search time grows roughly as O(n): When the data size increases by 10 times, the search time increases by approximately 10.
