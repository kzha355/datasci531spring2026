**Sensitivity and Conditioning**

<font size = "4">

- A **well-conditioned** problem is one where a small relative change in the input data causes a "reasonably small" relative change in the output.

- An **ill-conditioned** problem is one where a small relative change in the input data can cause a relative change in the output that is much larger.

- To quantify this, define the **condition number** as the ratio of the relative change in the solution to the relative change in the input.

- Suppose we are computing a single number $y = f(x)$, where $x, y \neq 0$. Let $\hat{y} = f(\hat{x})$. Then:

\begin{align*}
    \textrm{cond}(f) = \lim_{\delta \to 0}\ \sup_{|\hat{x} - x|\leq \delta} \dfrac{|f(\hat{x}) - f(x)|\ /\ |f(x)| }{ |\hat{x} - x|\ /\ |x|} = \lim_{\delta \to 0}\ \sup_{|\hat{x} - x|\leq \delta} \dfrac{|\hat{y} - y|\ /\ |y| }{ |\hat{x} - x|\ /\ |x|}
\end{align*}

- If $f$ is differentiable, then this is just:

\begin{align*}
    \textrm{cond}(f) = \frac{|f'(x)|}{|f(x)|\ /\  |x|} = \frac{|x|\cdot |f'(x)|}{|f(x)|}
\end{align*}

- This only holds when **both** $x \neq 0$ and $y = f(x) \neq 0$. Otherwise we would need to define a condition number in terms of the ratios of **absolute change** rather than relative change.

<font size = "4">

**Examples**

- Problem: Compute $y = f(x) = \dfrac{x}{2}$ for $x\neq 0$

$$ \textrm{cond}(f) = \frac{|x|\cdot \frac{1}{2}}{|x| \cdot \frac{1}{2}} = 1$$

The problem is well-conditioned.

- Problem: Compute $y = f(x) = \sqrt{x}$ for $x > 0$.

$$ \textrm{cond}(f) = \frac{x\cdot \frac{1}{2\sqrt{x}}}{\sqrt{x}} = \frac{x}{2\sqrt{x}\cdot\sqrt{x}} = \frac{x}{2x} = \frac{1}{2}$$

The problem is well-conditioned.

- Problem: Compute $y = f(x) = \tan(x)$, for $0 < x < \dfrac{\pi}{2}$

$$\textrm{cond}(f) = \frac{x\cdot (1 + \tan^2(x))}{\tan(x)} = x\cdot \left(\frac{1}{\tan(x)} + \tan(x) \right)$$



```python
import numpy as np

print(np.pi/2)
```

```python
x_vals = [1.570, 1.5707, 1.57079, 1.570796, 1.5707963]

for x in x_vals:
    cond_f = x * (1/np.tan(x) + np.tan(x))

    print(f"condition number of tan(x) for x = {x}:", round(cond_f, 4))
    print()
```

<font size = "4">

Computing $y = \tan(x)$ is very ill-conditioned for $x \approx \dfrac{\pi}{2}$

**Condition Numbers and Backward Stability**

<font size = "4">

- Suppose we want to calculate $y = f(x)$, but compute $\hat{y} = f(\hat{x})$ instead. (With $x, y\neq 0$)

- The condition number involves a limit, but if $|\hat{x} - x|$ is small, then:

\begin{align*}
    \textrm{cond}(f) \approx \dfrac{|\hat{y} - y|\ /\ |y| }{ |\hat{x} - x|\ /\ |x|}
\end{align*}

- This means:

$$ |\textrm{relative forward error}| \approx \textrm{cond}(f) \times  |\textrm{relative backward error}|$$

- **If we apply a backward stable algorithm to a well-conditioned problem, we will get an accurate solution**

- Or backward-stability + well-conditioned problem = forward-stability

- Because the input inevitably has error, maybe that's the value you wanted anyway!

**Machine Epsilon**

- On a computer, a real number $x$ is truncated and represented by a floating point representation $\hat{x} = \textrm{fl}(x)$

- If we assume **no** error from measurement, modeling, etc., there may still be error in representing the input on a computer.

- Sometimes its suprising:

```python
x = 0.3
y = 0.1 + 0.1 + 0.1

print(x == y)
```

<font size = "4">

- On most computers, there is a number $\epsilon_{\textrm{machine}} \approx 10^{-16}$ such that

$$ \frac{|\hat{x} - x|}{|x|} \leq \epsilon_{\textrm{machine}} $$

- By default, we can trust approximately 16 decimal digits for double precision floating point numbers

- If $\textrm{cond}(f) = 10^{k}$ then at best, we can trust around $16 - k$ decimal digits in the computed solution $\hat{y}$, since:

$$ |\textrm{relative forward error}| \approx \underbrace{10^k}_{\textrm{condition number}} \times  \underbrace{10^{-16}}_{\textrm{rel. backward error}} = 10^{-(16 - k)}$$

```python
# machine epsilon for double precision floats
eps_double = np.finfo(np.float64).eps
print(f"Machine epsilon for double-precision: {eps_double}")

# machine epsilon for single precision floats
eps_single = np.finfo(np.float32).eps
print(f"Machine epsilon for single-precision: {eps_single}")
```
