<span style="color:red; font-size:30px"> Sources of Error in Computing</span>

<font size = "4">

**Before computation**:

- *Modeling error*: some features of the problem or system have been simiplified or omitted.

- *Empirical measurements*: Limited precision of laboratory measurements, limited sample size, random noise

- *Previous computations*: Data may have been produced by previous computation with their own errors.

**During the computation**:

- *Truncation or discretization error*: The mathematical or statistical model could be simplified, or some infinite process could be terminated to get a practical algorithm

- *Rounding error*: Real numbers must be represented to some finite amount of precision, so computation is inexact.

<font size = "4">

**Example**

We could compute the surface area of the Earth by measuring its radius $r$ and using the formula for the surface area of a sphere:

$$A = 4\pi r^2$$

Approximations/errors include:

- Earth is modeled as a sphere, but this is just an idealization (modeling error)

- An estimate for the radius $r \approx 6370\ \textrm{km}$ is based on empirical measurements and previous computations

- The value for $\pi$ is given by an infinite limiting process, which must be approximated by a finite sum (truncation error) 

- The computations (exponentiation and multiplication) incur rounding error on a computer.

<font size = "4">

**Example**

If we want to compute the probability that a standard normal random variable is bounded by $x$, we need to compute the following integral:

$$P(X \leq x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^x \mathrm{e}^{-\frac{t^2}{2}}\ \textrm{d}t$$

The limiting processes that must be truncated are:

- The value of $\pi$.

- The value of $\sqrt{2\pi}$

- The unbounded lower limit of integration.

- The integral itself (no closed-form solution exists for arbitrary $x$).

Nevertheless, this can be computed (approximately) in many libraries:


```python
from scipy.stats import norm

x = 1.96

prob = norm.cdf(x)
print(prob)
```

<span style="color:red; font-size:30px"> Absolute Error and Relative Error</span>

<font size = "4">


- $\textrm{absolute error} = \textrm{approximate value} - \textrm{true value}$


- $\textrm{relative error} = \dfrac{\textrm{absolute error}}{\textrm{true value}} = \dfrac{\textrm{approximate value} - \textrm{true value}}{\textrm{true value}}$

<br>

Usually denote the true solution as $y$, and the computed solution as $\hat{y}$. Write the errors as:

- $E \coloneqq \textrm{absolute error} = \hat{y} - y$

- $\textrm{relative error} = \dfrac{\hat{y} - y}{y} = \dfrac{E}{y}$

- Relationship between the computed solution and true solution: $\hat{y} = y \cdot (1 + \textrm{relative error})$


<font size = "4">

We often are interested in the **magnitude** of the errors:

- $|\textrm{absolute error}| = |E| = |\hat{y} - y|$

- $|\textrm{relative error}| = \dfrac{|E|}{|y|} = \dfrac{|\hat{y} - y|}{|y|}$

Some observations about relative error:

- Can express relative error as a percentage. If $\dfrac{|E|}{|y|} = 0.1$, then we have 10% error.

- Rule of thumb: If $\dfrac{|E|}{|y|} \approx 10^{-p}$, then $\hat{y}$ has around $p$ correct significant decimal digits.

<span style="color:red; font-size:30px"> Data Error and Computational Error</span>

<font size = "4">

The true solution $y$ is related to the (true) input $x$ by some function: $y = f(x)$

- **Data error:** Sometimes we have the wrong input. $\hat{x}$ instead of $x$

- **Computational error:** We don't always compute $f$, but compute an *approximation* $\hat{f}$ instead.

So we actually compute $\hat{y} = \hat{f}(\hat{x})$

We can decompose the error as follows:

\begin{align*}
    E = \hat{y} - y &= \hat{f}(\hat{x}) - f(x)\\
    &= \underbrace{\hat{f}(\hat{x}) - f(\hat{x})}_{\textrm{computational error}} + \underbrace{f(\hat{x}) - f(x)}_{\textrm{propagated data error}}
\end{align*}

The magnitude can be bounded by the triangle inequality $|u + v| \leq |u| + |v|$

$$ |E| \leq |\hat{f}(\hat{x}) - f(\hat{x})| + |f(\hat{x}) - f(x)|$$

One term is controlled by the accuracy of the input, the other by the accuracy of the computational method.

<font size = "4">

**Simple Example**

Problem: Compute $\sin(\pi / 8)$

- True input: $x = \dfrac{\pi}{8}$

- Approximate input: use "biblical" approximation $\pi \approx 3$ so that $\hat{x} = \dfrac{3}{8}$

- Can write $f(x) = \sin(x)$ as $\sin(x) = x - \dfrac{x^3}{3!} + \dfrac{x^5}{5!} - \dfrac{x^7}{7!} + \cdots$ We'll truncate after 1 term, so $\hat{f}(x) = x$

We'll assume that ``np.pi`` and ``np.sin`` are "exact". We'll only look at the values to four decimal digits, so this is o.k. to assume for our purposes

```python
import numpy as np

x = np.pi/8
x_hat = 3/8

f = lambda x: np.sin(x)

f_hat = lambda x: x

y = f(x)
y_hat = f_hat(x_hat)

total_error = y_hat - y
comp_error = f_hat(x_hat) - f(x_hat)
data_error = f(x_hat) - f(x)

print("Computational error (4 digits):", round(comp_error, 4))
print("Propagated data error (4 digits):", round(data_error, 4))
print("Sum of errors:", round(comp_error + data_error, 4))
print("Total error (4 digits):", round(total_error, 4))
```

<font size = "4">

When looking at the "signed" error, we can get advantageous cancellation if the computational error and propagated data error have different signs.

When looking at magnitudes, the bound on the error can be pessimistic:

```python
E_abs = abs(total_error)
upper_bound = abs(comp_error) + abs(data_error)

print("Upper bound of error magnitude (4 digits):", round(upper_bound, 4))
print("Error magnitude (4 digits):", round(E_abs, 4))

print()

rel_error = E_abs / abs(y)

print("Relative error:", round(rel_error, 4), "or", str(round(rel_error*100,2)) + "%")
```

<span style="color:red; font-size:30px"> Truncation Error and Rounding Error</span>

<font size = "4">

We'll switch our focus to just computational error: truncation error and rounding.

Reducing one of these types of error can often come at the cost of the other, and a balance must be found.

**Example**

Problem: Compute $\dfrac{\textrm{d}}{\textrm{d}x} \sin(x)\bigg\rvert_{x=1}$ We know that the answer is $y = \cos(1) \approx 0.5403$

- We can approximate the solution by choosing a small value of $h > 0$ and using the finite difference formula

$$f'(x) \approx \frac{f(x + h) - f(x)}{h} =: \hat{y}$$

- By Taylor's theorem, there exists $c\in [x, x+h]$ such that:

$$f(x + h) = f(x) + f'(x)h + f''(c)\frac{h^2}{2}$$

- So if we had no rounding error, we would have the following result (below, $M$ is an upper bound for $|f''(c)|$ for values of $c$ near $x$)

$$ |\hat{y} - y| = |f''(c)| \frac{h}{2} \leq \frac{Mh}{2}$$

- It appears that by taking smaller and smaller values of $h$, we will get better and better answers.

- However, rounding error also affects the accuracy of the solution. Let $M$ be the bound on $|f''(c)|$ near $x$, and let $\epsilon > 0$ denote an upper bound on the relative error due to rounding on our computer (for modern computers using double precision arithmetic, $\epsilon \approx 10^{-16}$)

- Then an upper bound on the error (magnitude) is given by:

$$|\hat{y} - y| \leq \frac{Mh}{2} + \frac{2\epsilon}{h}$$

<span style="color:red; font-size:30px"> Forward Error and Backward Error</span>

<font size = "4">

Problem: compute $y = f(x)$. 

- We'll continue just focusing on computational error (so we will assume no input error is being made).

- The approximate solution is then $\hat{y} = \hat{f}(x)$, where $\hat{f}$ is the approximation to $f$ implemented on a computer.

- **Forward error**: $\Delta y \coloneqq \hat{y} - y$ (just the absolute error from above).

- Let $\hat{x}$ be the real number satisfying:

$$f(\hat{x}) = \hat{f}(x) = \hat{y}$$

- What does this mean? Our computed answer is what we would have obtained with the exact $f$ for a "perturbed" input $\hat{x}$.

- **Backward error**: $\Delta x \coloneqq \hat{x} - x$ where $f(\hat{x}) = \hat{f}(x)$.

<font size = "4">

From the perspective of **forward error analysis**, an algorithm is *stable* if the accumulated errors made by truncation and rounding are "small". All the focus is on the computational error.

From the perspective of **bacward error analysis**, an algorithm is *stable* if it computes the exact answer to a "nearby" problem.

Backward error analysis is:

- Often easier to carry out

- Allows us to explain **all** error in terms of the input.

- Fits well with the conditioning of a problem, which we cover in the next notebook.


<font size = "4">

**Example**

- $y = \sqrt{2} = 1.4142135...$

- $\hat{y} = 1.4$

```python
# Example

x = 2
y = np.sqrt(x)
y_hat = 1.4

x_hat = 1.96

print("x_hat is:", x_hat)
print("Square root of x_hat is:", np.sqrt(x_hat))

print()

print("Forward error, y_hat - y: ", np.round(abs(y_hat - y), 3))
print("Backward error, x_hat - x:", np.round(abs(x_hat - x), 3))
```

<font size = "4">

**Example**

- $f(x) = \cos(x)$

- $\hat{f}(x) = 1 - \dfrac{x^2}{2}$ (first two terms of Taylor Series)

- $y = \cos(1)$ (so $x = 1$)

- $\hat{y} = \hat{f}(x) = \hat{f}(1) = \dfrac{1}{2}$

```python
x = 1

y = np.cos(x)

y_hat = 1 - x**2 / 2

x_hat = np.acos(y_hat)

print("x_hat is:", round(x_hat, 4))
print("cos(x_hat) is:", round(np.cos(x_hat), 4)) # not actually exact!!

print()

print("Forward error, y_hat - y: ", np.round(abs(y_hat - y), 4))
print("Backward error, x_hat - x:", np.round(abs(x_hat - x), 4))
```
