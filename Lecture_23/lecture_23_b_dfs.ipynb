**General depth-first search**

- The knight's tour is a special case of a general depth-first search.

- In knight's tour, we try to create the deepest possible tree with *no* branches. If we get to a dead-end, we have to backtrack and try again, possibly many times.

- In a general **depth-first search (DFS)**, the goal is to search as deeply as possible, connecting as many nodes as possible, but branching where necessary.

- In fact, the DFS algorithm can create a group of disconnected trees, collectively known as a **depth-first forest**.

- Two new attributes must be added to the `Vertex` class, the **discovery time** and the **closing time**.

- The discovery time tracks the number of the steps in the algorithm before a vertex is first encountered.

- The closing time tracks the number of steps in the algorithm before a vertex is colored black.

**Implementation**

- You can either use `pip` to install `pythonds3` or just use the "adjacency_graph.py" I provided.

- Below, we define a `DFSGraph`, which is a subclass of the `Graph` class.

```python
from adjacency_graph import Graph, Vertex

class DFSGraph(Graph):
    def __init__(self):
        super().__init__()
        self.time = 0

    def dfs(self):
        for vertex in self:
            vertex.color = "white"
            vertex.previous = -1
        for vertex in self:
            if vertex.color == "white":
                self.dfs_visit(vertex)

    def dfs_visit(self, start_vertex):
        start_vertex.color = "gray"
        self.time = self.time + 1
        start_vertex.discovery_time = self.time
        for next_vertex in start_vertex.get_neighbors():
            if next_vertex.color == "white":
                next_vertex.previous = start_vertex
                self.dfs_visit(next_vertex)
        start_vertex.color = "black"
        self.time = self.time + 1
        start_vertex.closing_time = self.time
```

<font size = "4">

- This adds an extra attribute called `time`, which is used to define the discovery time and the closing time of a vertex.

- It creates two methods, which are used to complete the depth-first search.

- It proceeds as follows:

    1. Iterates over all the vertices in the graph, calling `dfs_visit` on all white-colored vertices.

    2. For each white vertex, it is colored gray and its `discovery_time` is updated to reflect the current value of the graph `time`.

    3. Then we search over all white-colored neighbors, recursively calling `dfs_visit` for each neighboring white node. It also updates the neighbors `previous` atrribute so they point to the current vertex.

    4. Once we have completed the search over the neighbors, the vertex is colored black and the `closing_time` is updated to reflect the graph's `time`.

- Recall that breadth-first search uses a `Queue`. This algorithm *implicitly* uses a Stack, the call stack used for recursive functions.

```python
# check the neighbors are in the right order
for k in G.get_vertices():
    print(k + ":")
    for nbr in G.get_vertex(k).get_neighbors():
        print(nbr)
    print()
```

```python
# Run the algorithm and check the vertices
G.dfs()

for k in G.get_vertices():
    print(G.get_vertex(k))
```

The discovery and closing times for the nodes in the DFS forest have a property called the **parenthesis property**. 

All the children of a node have a later discovery time, but an earlier closing time.

<font size = "4">

**Running time**

- Both loops in `dfs` are executed once for each vertex in the graph. Not counting what happens in `dfs_visit`, this has a $\mathcal{O}(|V|)$ cost.

- In `dfs_visit`, the recursive call happens only if the neighboring vertex is white. So at maximum, it runs once for every edge in the graph. This leads to $\mathcal{O}(|E|)$ cost.

- The total time is $\mathcal{O}(|V| + |E|)$
