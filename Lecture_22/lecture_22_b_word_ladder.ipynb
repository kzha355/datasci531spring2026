**The Word Ladder Problem**

- Consider the problem of transforming the word "fool" into "sage". However you must follow these rules:

    1. You can only change one letter at a time.

    2. At each step you must transform a word into another word. For example, you cannot transform "fool" into "fogl"

- The 2nd restriction is imposed to make it "fun" (by the problem's inventor Lewis Carroll), but transforming a string into another string with the fewest number of single-character edits has applications in linguistics and bioinformatics

- Our goal will be to find the smallest number of single-character transformations required to complete the transformation.

- One possible sequence we could do is:

    - fool
    - pool
    - poll
    - pole
    - pale
    - sale
    - sage


<font size = "4">

- We will solve this problem using a graph algorithm called **Breadth-First Search**. But first we need a graph.

- We would like to take a huge collection of 4-letter words that include both "fool" and "sage" and represent each word as a vertex. If two words differ by only a single letter in one location, then they will be connected by an edge.

- We'll load in a manageable number of 4-letter words stored in the file "words.txt"

Let's define the `Vertex` and `Graph` classes:

```python
class Vertex:
    def __init__(self, key, val=None):
        self.key = key 
        self.neighbors = {}
        self.val = val
    
    def get_neighbor(self, other_key):
        return self.neighbors.get(other_key) # fixed typo
    
    def set_neighbor(self, other_key, weight = 0):
        self.neighbors[other_key] = weight
    
    def __repr__(self):
        return f"Vertex({self.key})"
    
    def __str__(self):
        return str(self.key) + " connected to: " + str([x.key for x in self.neighbors])

    def get_neighbors(self):
        return self.neighbors.keys()

    def get_key(self):
        return self.key

class Graph:
    def __init__(self):
        self.vertices = {}
    
    def set_vertex(self, key):
        self.vertices[key] = Vertex(key)

    def get_vertex(self, key):
        return self.vertices.get(key)

    def __contains__(self, key):
        return key in self.vertices

    def add_edge(self, from_vert, to_vert, weight=0):
        if from_vert not in self.vertices:
            self.set_vertex(from_vert)
        if to_vert not in self.vertices:
            self.set_vertex(to_vert)
        self.vertices[from_vert].set_neighbor(self.vertices[to_vert], weight)

    def get_vertices(self):
        return self.vertices.keys()

    def __iter__(self):
        return iter(self.vertices.values())
```

Then we will create a graph, adding edges between words that are in the same bucket: 

```python
the_graph = Graph()

for word_group in buckets.values():
    for word1 in word_group:
        for word2 in word_group - {word1}:
            the_graph.add_edge(word1, word2)
```

Finally, we can wrap everything up in a function:

```python
def build_graph(filename):
    buckets = {}
    with open(filename, "r", encoding="utf8") as file_in:
        all_words = file_in.readlines()
    # create buckets of words that differ by 1 letter
    for line in all_words:
        word = line.strip()
        for i, _ in enumerate(word):
            bucket = f"{word[:i]}_{word[i + 1 :]}"
            buckets.setdefault(bucket, set()).add(word)

    the_graph = Graph()
    # add edges between different words in the same bucket
    for similar_words in buckets.values():
        for word1 in similar_words:
            for word2 in similar_words - {word1}:
                the_graph.add_edge(word1, word2)
    return the_graph
```
