### **Sorting Algorithms**

- In order to perform ordered sequential search or binary search, the list must be **ordered**.

- This requires us to sort the list first.

- When analyzing computational cost of a sorting algorithm, there are two relevant operations:

    - The number of **comparisons**
    - The number of **exchanges**



#### Bubble Sort

- Pass through the list, comparing consecutive elements.

- If element $i+1$ is less than element $i$, interchange their position in the list

- For a list of length $n$, the first pass through the list requires $n-1$ comparisons.

- After one pass, the largest element in the list will be moved to the end.

- The second pass requires $n-2$ comparisons. The third pass requires $n-3$ comparisons, etc.

```python
def bubble_sort(lst):
    n = len(lst)
    for p in range(n-1):
        for i in range(n-p-1):
            if lst[i] > lst[i+1]:
                tmp = lst[i+1]
                lst[i+1] = lst[i]
                lst[i] = tmp
            

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
bubble_sort(lst)
print(lst)
```

```python
def bubble_sort_easy_swap(lst):
    n = len(lst)
    for p in range(n-1):
        for i in range(n-p-1):
            if lst[i] > lst[i+1]:
                lst[i], lst[i+1] = lst[i+1], lst[i]

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
bubble_sort_easy_swap(lst)
print(lst)
```

- One more version (which is consistent with the text's implementation):

```python
def bubble_sort_v3(lst):
    n = len(lst)
    for p in range(n-1, 0, -1):
        for i in range(p):
            if lst[i] > lst[i+1]:
                lst[i], lst[i+1] = lst[i+1], lst[i]

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
bubble_sort_v3(lst)
print(lst)
```

#### Analysis of Bubble Sort

- Suppose we have a list of length $n$.

- $n-1$ passes will be made through the list. For pass $i$, $1 \leq i \leq n-1$, there will be $n - i$ comparisons.

- Total number of comparisons:

$$\sum_{i=1}^{n-1} (n - i) = \sum_{j=1}^{n-1}j = \frac{(n-1)n}{2} = \frac{n^2}{2} - \frac{n}{2} = \mathcal{O}(n^2)$$

- Number of exchanges:

    - Best case: list is already ordered. No exchanges.

    - Worst case: list is sorted in reverse order. Every comparison causes an exchange, so there are $\frac{n^2}{2} - \frac{n}{2} = \mathcal{O}(n^2)$ exchanges.

- Total cost:

    - Best case: $\mathcal{O}(n^2) + 0 = \mathcal{O}(n^2)$
    - Worse case: $\mathcal{O}(n^2) + \mathcal{O}(n^2) = \mathcal{O}(n^2)$



#### "Short Bubble"

- Bubble sort is widely considered the most inefficient sorting method.

- It does have one advantage however. If we complete a pass without making any exchanges, then the list **must already be sorted**.

- With a small modification, we can stop the algorithm early if it detects an ordered list.

```python
def short_bubble(lst):
    n = len(lst)
    for p in range(n-1):
        exchange = False
        for i in range(n-p-1):
            if lst[i] > lst[i+1]:
                lst[i], lst[i+1] = lst[i+1], lst[i]
                exchange = True 

        if not exchange:
            return None

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
short_bubble(lst)
print(lst)
```
