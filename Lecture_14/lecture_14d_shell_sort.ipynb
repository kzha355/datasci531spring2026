### **Shell Sort**

<font size = "3">

- Also known as the *diminishing increment sort*

- The algorithm was published by Donald Shell, so its name has nothing to do with shells.

- Breaks the original list into a set of smaller sublists, then each sublist is sorted using insertion sort.

- The Shell sort uses an increment $i$ known as the **gap**, to create the sublists by choosing all items that are $i$ apart in the list.

- The example below shows Shell sort applied to a list of $n = 9$ items with a gap of $i = 3$

- The way in which the sublist increments are chosen is a design choice.

- We need some decreasing sequence of increments/gaps $i_1 > i_2 > \cdots > i_m = 1$ to define the sublists which are sorted with insertion sort.

- The choice of sublist increments affect the performance of the algorithm in terms of computational complexity.

- In the following implementation, the first pass sorts $\lfloor\frac{n}{2}\rfloor$ sublists, the second $\lfloor\frac{n}{4}\rfloor$ sublists, then $\lfloor\frac{n}{8}\rfloor$ sublists, etc. until we reach $\lfloor \frac{n}{2^k}\rfloor = 1$ "sub"-list.

**Example: n = 19**

1. $\frac{19}{2} = 9 + \frac{1}{2}$. There are 9 sublists, defined using a gap of $i_1 = 9$. The first sublist has 3 elements, while the 8 other sublists have 2 elements

2. $\frac{19}{4} = 4 + \frac{3}{4}$. There are 4 sublists, defined using a gap of $i_2 = 4$. The first 3 sublists have 5 elements, while the last sublist has 4 elements.

3. $\frac{19}{8} = 2 + \frac{3}{8}$. There are 2 sublists, defined using a gap of $i_3 = 2$. The first sublist has 10 elements, the second sublist has 9 elements

3. $\frac{19}{16} = 1 + \frac{3}{16}$. There is 1 sublist, defined using a gap of $i_4 = 1$. The sublist has 19 elements.

```python
def shell_sort(a_list):
    sublist_count = len(a_list) // 2
    while sublist_count > 0:
        for pos_start in range(sublist_count):
            gap_insertion_sort(a_list, pos_start, sublist_count)
        print("After increments of size", sublist_count, "the list is", a_list)
        sublist_count = sublist_count // 2


def gap_insertion_sort(a_list, start, gap):
    for i in range(start + gap, len(a_list), gap):
        cur_val = a_list[i]
        cur_pos = i
        while cur_pos >= gap and a_list[cur_pos - gap] > cur_val:
            a_list[cur_pos] = a_list[cur_pos - gap]
            cur_pos = cur_pos - gap
        a_list[cur_pos] = cur_val


import random 

lst = random.sample(range(50), k=19)
print(lst)
shell_sort(lst)
```

<font size = "3">

Here is an implementation that only uses `for` loops.

```python
import numpy as np 

def my_shell_sort(lst):
    n = len(lst)
    passes = int(np.floor(np.log2(n)))
    sublist_ct = n // 2
    for sweep in range(passes):
        for j in range(sublist_ct):
            my_gap_insertion_sort(lst, j, sublist_ct)
        print("After increments of size", sublist_ct, "the list is", lst)
        sublist_ct = sublist_ct // 2


def my_gap_insertion_sort(lst, start, gap):
    n = len(lst)

    for p in range(start + gap, n, gap):
        next_val = lst[p]
        insert = start
        for pos in range(p, 0, -gap):
            if lst[pos-gap] <= next_val:
                insert = pos
                break
            lst[pos] = lst[pos-gap]
        lst[insert] = next_val

lst = [4, 44, 28, 33, 34, 32, 21, 3, 41, 27, 14, 5, 7, 23, 26, 18, 0, 37, 20]
my_shell_sort(lst)
```

<font size = "3">

- Can Shell sort be better than insertion sort? Don't we do a complete pass of insertion sort at the end?

- The final insertion sort doesn't need to do as many shifts, since the list has been partially resorted.

- You could also optimize the code to avoid unnecessary comparisons based on the knowledge of the presorted elements.

- The performance of Shell sort depends heavily on the choice of gap sequence.

- **Worst case gap sequence:** $\mathcal{O}(n^2)$

- **Best-case for "most" gap sequences:** $\mathcal{O}(n\log n)$

- **Average case:** Highly dependent on gap sequence.
