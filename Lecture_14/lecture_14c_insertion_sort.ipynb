### **Insertion Sort**

<font size = "3">

- Maintains a sorted sublist in the lower positions of the list.

- Each iteration, the next item is inserted into the sublist such that it remains ordered:

The text does the following implementation with a `while` loop:


```python
def insertion_sort(a_list):
    for i in range(1, len(a_list)):
        cur_val = a_list[i]
        cur_pos = i

        while cur_pos > 0 and a_list[cur_pos - 1] > cur_val:
            a_list[cur_pos] = a_list[cur_pos - 1]
            cur_pos = cur_pos - 1
        a_list[cur_pos] = cur_val


a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print(a_list)
insertion_sort(a_list)
print(a_list)
```

<font size = "3">

Here's an implementation using only `for` loops:

```python
def my_insertion_sort(lst):
    n = len(lst)

    for p in range(1, n):
        next_val = lst[p]
        insert = 0
        for pos in range(p, 0, -1):
            if lst[pos-1] <= next_val:
                insert = pos
                break
            lst[pos] = lst[pos-1]
        lst[insert] = next_val

a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print(a_list)
my_insertion_sort(a_list)

print(a_list)
```

**Worst-case scenario**

- There are $n-1$ passes through the list

- For each pass $1 \leq p \leq n-1$

    - $p$ comparisons

    - $p+1$ shifts (assignments **into** the list)

- So the number of operations (comparisons + shifts) is:

$$\sum_{p=1}^{n-1}(p + p + 1) = (n-1) + 2\sum_{p=1}^{n-1}p = (n - 1) + \mathcal{O}(n^2) = \mathcal{O}(n^2)$$

**Best-case scenario**

- There are $n-1$ passes through the list

- For each pass $1 \leq p \leq n-1$

    - 1 comparison

    - 1 "shift"

- So the number of operations is:

$$\sum_{p=1}^{n-1}2 = 2(n-1) = \mathcal{O}(n)$$

**Average complexity**

- Check out [this Stack Overflow post](https://stackoverflow.com/questions/17055341/why-is-insertion-sort-%ce%98n2-in-the-average-case/17055342#17055342)

- The average cost is $\mathcal{O}(n^2)$


**Note:**

- A shift operation approximately requires only 1/3 the work of an exchange, since only one assignment is performed.

- Because of this, insertion sort has good performance on benchmark studies. Remember, the number of operations is not always the end of the story!
