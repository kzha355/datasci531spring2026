### **Selection Sort**

- Improvement on bubble sort - only one exchange is made for every pass through the list

- On each pass, look for the largest item. When pass is complete, place it in the proper location.

- Still requires $n-1$ passes to sort a list of length $n$

- The following function implements the algorithm:

```python
def selection_sort(lst):
    n = len(lst)
    for p in range(n-1):
        max_ind = None 
        max_val = float("-inf")
        for i in range(n-p):
            if lst[i] > max_val:
                max_ind = i 
                max_val = lst[i]
        if max_ind != n-p-1:
            lst[n-p-1], lst[max_ind] = lst[max_ind], lst[n-p-1]
    
lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
selection_sort(lst)
print(lst)
```

<font size = "3">

- Compare with the book's version

```python
def selection_sort_book(a_list):
    for i, item in enumerate(a_list):
        min_idx = len(a_list) - 1
        for j in range(i, len(a_list)):
            if a_list[j] < a_list[min_idx]:
                min_idx = j
        if min_idx != i:
            a_list[min_idx], a_list[i] = a_list[i], a_list[min_idx]

lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
selection_sort_book(lst)
print(lst)
```

<font size = "3">

- Selection sort makes the same number of comparisons as bubble sort. So its cost is also $\mathcal{O}(n^2)$

- However, it has fewer exchanges, so it outperforms bubble sort on benchmarks (even though they are both $\mathcal{O}(n^2)$)
