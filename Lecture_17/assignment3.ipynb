**(Q1)** The bubble sort algorithm can be modified to "bubble" in both directions. The first pass moves "up" the list, and the maximum element "bubbles" to the top. The second pass moves "down" the list, and the minimum element "bubbles" to the bottom. This alternating pattern continues until no more passes are necessary. 

Write a function that implements this algorithm.

````python
def round_bubble(lst):
    left = 0
    right = len(lst) - 1

    while left < right:
        exchange = False

        for i in range(left, right):
            if lst[i] > lst[i + 1]:
                lst[i], lst[i + 1] = lst[i + 1], lst[i]
                exchange = True
        right -= 1

        for i in range(right, left, -1):
            if lst[i] < lst[i - 1]:
                lst[i], lst[i - 1] = lst[i - 1], lst[i]
                exchange = True
        left += 1

        if not exchange:
            return None
````

````python
# TEST
lst = [59, 7, 8, 73, 20, 488, 82, 67, 73]
round_bubble(lst)
print(lst)
````

**(Q2)** The implementation of the merge sort algorithm we saw had two deficiencies:

1. It used **slicing** as follows:

```python
        left_half = a_list[:mid]
        right_half = a_list[mid:]
        merge_sort(left_half)
        merge_sort(right_half)
```
Each slicing operation incurs an additional $O(k)$ cost, where $k$ is the size of the slice.

2. Once the slicing operation is executed, we need extra memory to store the two halves of the list independently of the original list. This increases the algorithm's *space* complexity.

Implement a version of merge sort that does not use slicing and sorts the list **in-place**. Some starter code is provided below for you to complete.

What sorting algorithm should be used for merging two sorted sub-lists?

````python
def merge_sort(a_list):
    merge_sort_helper(a_list, 0, len(a_list))

def merge_sort_helper(lst, start_index, end_index):
    # 1. Base case of recursion
    if end_index - start_index <= 1:
        return

    # 2. Two recursive calls to merge_sort_helper
    mid = (start_index + end_index) // 2
    merge_sort_helper(lst, start_index, mid)
    merge_sort_helper(lst, mid, end_index)

    # 3. Merge the two halfs of the list (in-place!)
    while mid < end_index:
        val = lst[mid]
        j = mid - 1
        while j >= start_index and lst[j] > val:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = val
        mid += 1
````

````python
# TEST
lst = [59, 7, 8, 73, 20, 488, 82, 67, 73]
merge_sort(lst)
print(lst)
````

**(Q3)** One way to improve the quicksort algorithm is to enforce a "partition limit". Instead of treating $n = 1$ as the base case of the recursion, we choose a short length $m > 1$ for the base case. When the length of the sub-list is $\leq m$, we end the recursion and sort it directly using insertion sort.

- Implement this version of quicksort where the partition limit is included as an additional input parameter. For $m = 1$, it reduces to the original quicksort.

- Perform benchmark tests on large lists of randomly generated integers. Vary both the size of the lists ($n$) and the partition limit ($m$) and see how performance is affected, in terms of time.

````python
def quick_sort3(lst, m):
    quick_sort_helper3(lst, 0, len(lst) - 1, m)


def quick_sort_helper3(lst, first, last, m):
    if last - first + 1 <= m:
        insertion_sort(lst, first, last)
    elif first < last:
        split = partition(lst, first, last)
        quick_sort_helper3(lst, first, split - 1, m)
        quick_sort_helper3(lst, split + 1, last, m)


def insertion_sort(lst, first, last):
    for i in range(first + 1, last + 1):
        key = lst[i]
        j = i - 1
        while j >= first and lst[j] > key:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = key


def partition(lst, first, last):
    pivot_val = lst[first]
    left_mark = first + 1
    right_mark = last
    done = False

    while not done:
        while left_mark <= right_mark and lst[left_mark] <= pivot_val:
            left_mark = left_mark + 1
        while left_mark <= right_mark and lst[right_mark] >= pivot_val:
            right_mark = right_mark - 1
        if right_mark < left_mark:
            done = True
        else:
            lst[left_mark], lst[right_mark] = lst[right_mark], lst[left_mark]

    lst[first], lst[right_mark] = lst[right_mark], lst[first]
    return right_mark
````

````python
# TEST
lst = [59, 7, 8, 73, 20, 488, 82, 67, 73]
quick_sort3(lst, 2)
print(lst)
````

````python
import random
import time

n_values = [10, 100, 1000]
m_values = [5, 10]

for n in n_values:
    for m in m_values:
        lst = [random.randint(0, 100000) for _ in range(n)]
        t0 = time.time()
        quick_sort3(lst, m)
        t1 = time.time()
        print(f"n={n}, m={m}: {(t1-t0)*1000} ms")
````

**(Q4)**

1. Implement a variant of quicksort where the **last** element of the list is used as the pivot value.

2. Implement a variant of quicksort where the **middle** element of the list is used as the pivot value.

````python
# last element as pivot value
def quick_sort4(a_list):
    quick_sort_helper4(a_list, 0, len(a_list) - 1)


def quick_sort_helper4(a_list, first, last):
    if first < last:
        split = partition(a_list, first, last)
        quick_sort_helper4(a_list, first, split - 1)
        quick_sort_helper4(a_list, split + 1, last)


def partition(a_list, first, last):
    pivot_val = a_list[last]
    left_mark = first
    right_mark = last - 1
    done = False

    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark = left_mark + 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark = right_mark - 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]

    a_list[last], a_list[left_mark] = a_list[left_mark], a_list[last]
    return left_mark
````

````python
# TEST
lst = [59, 7, 8, 73, 20, 488, 82, 67, 73]
quick_sort4(lst)
print(lst)
````

````python
# middle element as pivot value
def quick_sort(a_list):
    quick_sort_helper(a_list, 0, len(a_list) - 1)


def quick_sort_helper(a_list, first, last):
    if first < last:
        split = partition(a_list, first, last)
        quick_sort_helper(a_list, first, split)
        quick_sort_helper(a_list, split + 1, last)


def partition(a_list, first, last):
    pivot_val = a_list[(first+last)//2]
    left_mark = first
    right_mark = last
    done = False

    while not done:
        while left_mark <= right_mark and a_list[left_mark] < pivot_val:
            left_mark = left_mark + 1
        while left_mark <= right_mark and a_list[right_mark] > pivot_val:
            right_mark = right_mark - 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]
            left_mark = left_mark + 1
            right_mark = right_mark - 1

    return right_mark
````

````python
# TEST
lst = [59, 7, 8, 73, 20, 488, 82, 67, 73]
quick_sort(lst)
print(lst)
````

**(Q5)** Implement quicksort where the "median of three" method is used for selecting the pivot. That is, compare the first, middle, and last elements of a list and use the median as the pivot value.

````python
def median_sort(a_list, first, last):
    mid = (first + last) // 2
    if a_list[first] > a_list[mid]:
        a_list[first], a_list[mid] = a_list[mid], a_list[first]
    if a_list[first] > a_list[last]:
        a_list[first], a_list[last] = a_list[last], a_list[first]
    if a_list[mid] > a_list[last]:
        a_list[mid], a_list[last] = a_list[last], a_list[mid]
    a_list[first], a_list[mid] = a_list[mid], a_list[first]

def quick_sort5(a_list):
    quick_sort_helper5(a_list, 0, len(a_list) - 1)

def quick_sort_helper5(a_list, first, last):
    if first < last:
        split = partition(a_list, first, last)
        quick_sort_helper5(a_list, first, split - 1)
        quick_sort_helper5(a_list, split + 1, last)

def partition(a_list, first, last):
    median_sort(a_list, first, last)
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False

    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark += 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark -= 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = a_list[right_mark], a_list[left_mark]

    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]
    return right_mark
````
