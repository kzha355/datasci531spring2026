### Some vocabulary needed to define trees

<font size = "3">

- **Node:** Fundamental part of a tree. Can have a name, known as a *key*. Node can contain additional info called the *value* or *payload*. The payload info is not central to many tree algorithms, but is often critical in applications using trees.

- **Edge:** Fundamental part of a tree. Connects two nodes to show there is a relationship between them. Every non-root node is connected by exactly one incoming edge from another node. Each node may have several outgoing edges.

- **Root:** The only node with no incoming edges.

- **Path:** Ordered list of nodes that are connected by edges. E.g. Mammalia --> Carnivora --> Felidae --> *Felis* --> *catus* is a path.

- **Children:** Set of nodes that have incoming edges from the same node are children of that node.

- **Parent:** A node is the parent of all the nodes it connects to with outgoing edges.

- **Sibling:** Nodes in the tree that are children of the same parent.

- **Subtree:** A set of nodes and edges consisting of a parent and all descendants of that parent.

- **Leaf node:** Node with no children.

- **Level:** The level of a node is the number of edges on the path from the root to that node. The level of the root node is zero.

- **Height:** The maximum level of any node in the tree

<font size = "3">

**First definition of tree**


A tree consists of a set of nodes and a set of edges that connect pairs of nodes. A tree has the following properties:

- One node of the tree is designated as the root node.

- Every node $n$, except the root node, is connected by an edge from exactly one parent node.

- A unique path traverses from the root to each node.

- If each node in the tree has a maximum of two children, it is a **binary tree**

**Second (recursive) definition of tree**

- A tree is either empty or consists of a root and zero or more subtrees, each of which is a tree.

- The root of each subtree is connected to the root of the parent tree by an edge.
