### Implementation of Binary Tree

- We will represent a binary tree using nodes and references.

- Actually, this will be implemented **recursively**. Instead of adding individual nodes, a tree will point to other subtrees.

- The class `BinaryTree` below refers to itself.

#### Some notes on the implementation

- For simplicity, we'll ignore important design principles like which attributes should be private, or creating properties.

- When we initialize a BinaryTree, it will consist of a single node with a *key* (a label or value), and a *left child* and *right child* initialized to `None`

- To insert a left child node, we have to consider two cases: whether the current left child is `None` or not.

- If it is `None`, assign a new `BinaryTree` object to the left child.

- From the `__init__` method, we know that this new `BinaryTree` is just a single node that has no children (doesn't point to anything)

- If the left child is not `None`, then it is another `BinaryTree` object (if the programmer using our class has been behaving themselves).

- Then we insert a new "Node" (really a new `BinaryTree`) as the new left child, and push the existing child down one level in the tree.

- A similar approach holds when we insert a right child node.

```python
class BinaryTree:
    def __init__(self, root_val):
        self.key = root_val
        self.left_child = None 
        self.right_child = None

    def insert_left(self, new_item):
        if self.left_child is None:
            self.left_child = BinaryTree(new_item)
        else:
            new_child = BinaryTree(new_item)
            new_child.left_child = self.left_child 
            self.left_child = new_child
    
    def insert_right(self, new_item):
        if self.right_child is None:
            self.right_child = BinaryTree(new_item)
        else:
            new_child = BinaryTree(new_item)
            new_child.right_child = self.right_child 
            self.right_child = new_child 

    def get_root_val(self):
        return self.key 

    def set_root_val(self, new_val):
        self.key = new_val 

    def get_left_child(self):
        return self.left_child 
    
    def get_right_child(self):
        return self.right_child

    def __str__(self):
        return f"BinaryTree with root value {self.key}"

    def __repr__(self):
        return f"BinaryTree with root value {self.key}"
```

Here is an example utilizing the `BinaryTree` class. What does the tree look like after each command?

```python
a_tree = BinaryTree("a")

print(a_tree.get_root_val())
print(a_tree.get_left_child())
a_tree.insert_left("b")
print(a_tree.get_left_child())
print(a_tree.get_left_child().get_root_val())
a_tree.insert_right("c")
print(a_tree.get_right_child())
print(a_tree.get_right_child().get_root_val())
a_tree.get_right_child().set_root_val("hello")
print(a_tree.get_right_child().get_root_val())
```
