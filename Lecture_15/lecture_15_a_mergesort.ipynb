### **Merge sort**

- This is a **divide-and-conquer** strategy to improve sorting algorithm performance.

- Base case: An empty list or a list with one item is sorted by definition.

- If a list has more than one item, split it in two and recursively merge sort both halves.

- Once the two halves are sorted, a **merge** is performed, which combines the two smaller sorted list into a single sorted new list.

- Below is a recursive implementation of merge sort adapted from the text.

- It includes print statements describing the splitting and merging steps.

````python
def merge_sort(a_list):
    if len(a_list) <= 1:
        return 

    print("Splitting", a_list)
    mid = len(a_list) // 2
    left_half = a_list[:mid]
    right_half = a_list[mid:]

    # recursive call to each half
    merge_sort(left_half)
    merge_sort(right_half)
    print("Merging",left_half,"and",right_half)

    # Merging process
    i, j, k = 0, 0, 0
    while i < len(left_half) and j < len(right_half):
        if left_half[i] <= right_half[j]:
            a_list[k] = left_half[i]
            i = i + 1
        else:
            a_list[k] = right_half[j]
            j = j + 1
        k = k + 1

    while i < len(left_half):
        a_list[k] = left_half[i]
        i = i + 1
        k = k + 1

    while j < len(right_half):
        a_list[k] = right_half[j]
        j = j + 1
        k = k + 1
    
````

````python
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
merge_sort(a_list)
print(a_list)
````

- I wanted to see if I could write a version using only for loops, without writing an excessive amount of `if` statements.

- The merging process in the implementation is quite compact, and requires no `if` statements (it does come at the cost of defining an additional function called `check_max`)

- It demonstrates how you can use the `max` function in a more general way

- I also wrote it to avoid calling `.pop(0)` on a list - the cost of data structure methods is an important consideration in algorithm design!

````python
def check_max(half):
    if len(half) == 0:
        return float('-inf')
    return half[-1]

def my_merge_sort(a_list):
    if len(a_list) <= 1:
        return 

    print("Splitting", a_list)
    mid = len(a_list) // 2
    left_half = a_list[:mid]
    right_half = a_list[mid:]

    # recursive call to each half
    my_merge_sort(left_half)
    my_merge_sort(right_half)
    print("Merging",left_half,"and",right_half)

    # Merging process
    halves = [left_half, right_half]

    f = lambda ind: check_max(halves[ind])

    for k in range(len(a_list) - 1, -1, -1):
        i = max([0, 1], key = f)
        a_list[k] = halves[i].pop()
````

````python
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
my_merge_sort(a_list)
print(a_list)
````

- A sorting algorithm is **stable** if two objects are considered "equal", then after sorting, they appear in the same order as the original list.

- When sorting more complex objects, this can be important if the list is sorted multiple times with different criteria.

- For example if we want to sort a list containing tuples of `(name, age)` pairs, we could sort them so the names are in alphabetical order, then sort by age.

- If a stable sorting algorithm is used, then the list would be ordered youngest to oldest, and people with the same age would appear in the list alphabetically. We could not guarantee that if we used an unstable sorting algorithm.

### Cost of Merge Sort

<font size = "3">

- Let $n$ be the length of the list. For convenience, assume $n = 2^k$

- We must analyze two distinct processes: splitting and merging.

- From binary search we saw that we can divide the list in half $k = \log_2(n)$ times. 

- At each level of the recursion, a merge must be performed.

    - At the lowest level of the recursion, $n = 2^k$ lists of size 1 must be merged. Each of the $n$ items will need to be assigned to the appropriate slot of a list of size 2. So there are $\mathcal{O}(n)$ operations.

    - At the 2nd-lowest level, $\frac{n}{2} = 2^{k-1}$ lists of size 2 must be merged. There are a total of $n$ items across all the lists. Each of the $n$ items will need to be assigned to the appropriate slot of a list of size 4. So there are $\mathcal{O}(n)$ operations.

    - At the 3rd-lowest level, $\frac{n}{4} = 2^{k-2}$ lists of size 4 must be merged. Once again, there are $\mathcal{O}(n)$ operations.

    - This pattern continues until the highest level of the recursion, where $2 = 2^{k-(k-1)}$ lists of size $\frac{n}{2}$ must be merged. Once again, there are $\mathcal{O}(n)$ operations.

- There are $\log_2(n)$ levels, each of which incurs a cost of $\mathcal{O}(n)$. 

- So the total cost is $\mathcal{O}(n\log(n))$

- This is the **worst case** cost.



### Some caveats

<font size = "3">

- In Python, the list slicing operation `lst[:m]` is an $\mathcal{O}(m)$ operation.

- To ensure total cost is $\mathcal{O}(n\log(n))$, we must change the algorithm to avoid slicing, by passing starting and ending indices to the recursive call instead.

- The reduction in computational complexity comes at the cost of **increased space complexity**. Extra memory is required to hold the two halves of the list.

- This extra space complexity can be prohibitive for very large lists or datasets.
