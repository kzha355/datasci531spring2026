### **Quicksort**

<font size = "3">

- This is another **divide-and-conquer** sorting algorithm.

- Unlike merge sort, the division of the list may not be equal. This can hurt performance.

- The algorithm starts by selecting a **pivot value**. There are a variety of strategies for choosing it. The most simple approach is to choose the first element

- For an example, we will sort the following list using quicksort with the first element as the pivot value:

  <img src="files/firstsplit.png" alt="Centered image" width = "400">

</div>

<font size = "3">

- The algorithm then determines where in the list the pivot value *should* be. This is the **split point**.

- In the process of determining the split point, we move all items $\leq$ the pivot value in the list to the left. All items $\geq$ the pivot value are moved right.

- The value located at the split point is then swapped with the pivot value.

- The process involves two **position markers**, we'll call them `left_mark` and `right_mark`. They move towards each other moving over items that are on the "wrong" side, until they pass each other. The final value of `right_mark` determines the split point 

<div style="text-align: center;">
  <img src="files/partitionA.png" alt="Centered image" width = "400">
  <figcaption><font size = "1"> Miller, Randum, Yasinovskyy (Problem Solving with Algorithms and Data Structures using Python)</figcaption>
</div>


<font size = "3">

- Once the position markers cross, the right marker is determined to be the split point.

- Then we can recursively apply the algorithm to each sub-list to the left and right of the split point.

<div style="text-align: center;">
  <img src="files/partitionB.png" alt="Centered image" width = "400">
  <figcaption><font size = "1"> Miller, Randum, Yasinovskyy (Problem Solving with Algorithms and Data Structures using Python)</figcaption>
</div>

Here is the implementation of quick sort:

```python
def quick_sort(a_list):
    quick_sort_helper(a_list, 0, len(a_list) - 1)


def quick_sort_helper(a_list, first, last):
    if first < last:
        split = partition(a_list, first, last)
        quick_sort_helper(a_list, first, split - 1)
        quick_sort_helper(a_list, split + 1, last)


def partition(a_list, first, last):
    pivot_val = a_list[first]
    left_mark = first + 1
    right_mark = last
    done = False

    while not done:
        while left_mark <= right_mark and a_list[left_mark] <= pivot_val:
            left_mark = left_mark + 1
        while left_mark <= right_mark and a_list[right_mark] >= pivot_val:
            right_mark = right_mark - 1
        if right_mark < left_mark:
            done = True
        else:
            a_list[left_mark], a_list[right_mark] = (
                a_list[right_mark],
                a_list[left_mark],
            )
    a_list[first], a_list[right_mark] = a_list[right_mark], a_list[first]

    return right_mark
```

```python
lst = [54, 26, 93, 17, 77, 31, 44, 55, 20]
quick_sort(lst)
print(lst)
```

**Question:** Is quicksort a *stable* sorting algorithm?

The following `Pair` class will help us keep track of duplicate integers with an additional string label.

```python
class Pair:
    def __init__(self, num, label):
        self.data = (num, label)

    def __gt__(self, pair):
        return self.data[0] > pair.data[0]

    def __lt__(self, pair):
        return self.data[0] < pair.data[0]

    def __ge__(self, pair):
        return not self.__lt__(pair)
    
    def __le__(self, pair):
        return not self.__gt__(pair)

    def __str__(self):
        return str(self.data[0]) + self.data[1]

    def __repr__(self):
        return str(self.data[0]) + self.data[1]
```

```python
x = Pair(1, 'a')
y = Pair(1, 'b')
z = Pair(2, '')

print(x, y, z)
print(x <= y)
print(x >= y)
print(z > x)
```

### Analysis of Quick sort
<font size = "3">

- In the best case, the split point will be exactly in the middle of the list. Then it essentially boils down to mergesort (but without the extra memory requirements!).

- So the best case computational complexity is $\mathcal{O}(n\log(n))$.

- Performance suffers when the split points are far from the midpoint, skewed very far to the left or right.

- The worst case scenario is that every split point lands at the beginning or end of the list.

- Then the "recursive" algorithm reduces to sorting a list of size 0 and size $n-1$, then sorting a list of size 0 and size $n-2$, and so on.

- So the worst-case scenario is $\mathcal{O}(n^2)$, plus the overhead of unnecessary recursion (creating stack frames etc.)

### Choosing pivots

<font size = "3">

- The choice of pivot value has consequences for the distribution of the split points.

- It is ideal for the split points to be near the middle of the list. 

- The **median of three** approach compares the first element, last element, and middle element. The median of the three is chosen as the pivot.

- Another option is to choose a random pivot. There are also more sophisticated strategies for choosing the pivots.
